## Blocagem

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
algoritmos = [
    "PhonexStaticBlocking-50",
    "PhonexStaticBlocking-25",
    "PhonexStaticBlocking-15",
    "PhonexStaticBlocking-5",
    "SoundexBlocking"
]

tempos = [12, 8.6, 7.7, 7.7, 9.2]
precisao = [0.9629629629629629, 0.9617021276595744, 0.9694323144104804, 0.9613259668508287, 0.963855421686747]
recall = [0.782608695652174, 0.7558528428093646, 0.7424749163879598, 0.5819397993311036, 0.802675585284281]
medida_f = [0.8634686346863469, 0.846441947565543, 0.8409090909090908, 0.7249999999999999, 0.8759124087591241]

# Criar subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Gráfico de Tempo
axs[0, 0].bar(np.arange(len(algoritmos)), tempos, color='blue')
axs[0, 0].set_ylabel('Tempo (s)')
axs[0, 0].set_title('Tempo de Execução')
axs[0, 0].set_xticks(np.arange(len(algoritmos)))
axs[0, 0].set_xticklabels(algoritmos, rotation=15, ha="right")

# Gráfico de Precisão
axs[0, 1].bar(np.arange(len(algoritmos)), precisao, color='green')
axs[0, 1].set_ylabel('Precisão')
axs[0, 1].set_title('Precisão')
axs[0, 1].set_xticks(np.arange(len(algoritmos)))
axs[0, 1].set_xticklabels(algoritmos, rotation=15, ha="right")

# Gráfico de Recall
axs[1, 0].bar(np.arange(len(algoritmos)), recall, color='red')
axs[1, 0].set_ylabel('Recall')
axs[1, 0].set_title('Recall')
axs[1, 0].set_xticks(np.arange(len(algoritmos)))
axs[1, 0].set_xticklabels(algoritmos, rotation=15, ha="right")

# Gráfico de Medida-F
axs[1, 1].bar(np.arange(len(algoritmos)), medida_f, color='purple')
axs[1, 1].set_ylabel('Medida-F')
axs[1, 1].set_title('Medida-F')
axs[1, 1].set_xticks(np.arange(len(algoritmos)))
axs[1, 1].set_xticklabels(algoritmos, rotation=15, ha="right")

# Ajustar o layout
plt.tight_layout()

fig.suptitle('PhonexStaticBlocking vs SoundexBlocking', y=1)

# Exibir o gráfico
plt.show()
```

## Incremental x Estático

```python
def generate_incremental_vs_static_charts(tempos, precisao, recall, medida_f, title):
  algoritmos = [
    "Estático (100%)",
    "Incremental (100%)",
    "Estático Ingênuo (100%)"
  ]

  alg_tempo = ['Estático (100%)', 'Incremental (50%)', 'Incremental (35%)', 'Incremental (15%)']

  # Criar subplots
  fig, axs = plt.subplots(2, 2, figsize=(12, 8))

  # Gráfico de Tempo
  axs[0, 0].bar(np.arange(len(alg_tempo)), tempos[:-1], color='blue')
  axs[0, 0].set_ylabel('Tempo (s)')
  axs[0, 0].set_title('Tempo de Execução')
  axs[0, 0].set_xticks(np.arange(len(alg_tempo)))
  axs[0, 0].set_xticklabels(alg_tempo, rotation=15, ha="right")

  # Gráfico de Precisão
  axs[0, 1].bar(np.arange(len(algoritmos)), precisao, color='green')
  axs[0, 1].set_ylabel('Precisão')
  axs[0, 1].set_title('Precisão')
  axs[0, 1].set_xticks(np.arange(len(algoritmos)))
  axs[0, 1].set_xticklabels(algoritmos, rotation=15, ha="right")

  # Gráfico de Recall
  axs[1, 0].bar(np.arange(len(algoritmos)), recall, color='red')
  axs[1, 0].set_ylabel('Recall')
  axs[1, 0].set_title('Recall')
  axs[1, 0].set_xticks(np.arange(len(algoritmos)))
  axs[1, 0].set_xticklabels(algoritmos, rotation=15, ha="right")

  # Gráfico de Medida-F
  axs[1, 1].bar(np.arange(len(algoritmos)), medida_f, color='purple')
  axs[1, 1].set_ylabel('Medida-F')
  axs[1, 1].set_title('Medida-F')
  axs[1, 1].set_xticks(np.arange(len(algoritmos)))
  axs[1, 1].set_xticklabels(algoritmos, rotation=15, ha="right")

  # Ajustar o layout
  plt.tight_layout()

  fig.suptitle(title, y=1)

  # Exibir o gráfico
  plt.show()
```

#### CD Information 

```python
tempos = [8.7, 4.9, 4.2, 2.6, 1140]
precisao = [0.9638, 0.9638, 0.9440]
recall = [0.8026, 0.8026, 0.8461]
medida_f = [0.8759, 0.8759, 0.8924]

generate_incremental_vs_static_charts(tempos, precisao, recall, medida_f, 'Desempenho - CD Information')
```

## ProxCluster vs DuDe

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
algoritmos = ["ProxCluster", "DuDe"]
tempos = [8.6, 0.8]
precisao = [0.9638, 0.9664]
recall = [0.8026, 0.4816]
medida_f = [0.8759, 0.6428]

# Criar figura com dois subplots
fig, (ax1, ax2) = plt.subplots(2, figsize=(10, 8))

# Gráfico de Tempo
ax1.bar(algoritmos, tempos, color='blue')
ax1.set_ylabel('Tempo (s)')
ax1.set_title('Tempo de Execução')

# Gráfico de Precisão, Recall e Medida-F
bar_width = 0.2
index = np.arange(len(algoritmos))

ax2.bar(index, precisao, bar_width, label='Precisão', color='green')
ax2.bar(index + bar_width, recall, bar_width, label='Recall', color='red')
ax2.bar(index + 2 * bar_width, medida_f, bar_width, label='Medida-F', color='purple')

# Configuração dos eixos e rótulos
ax2.set_ylabel('Valores')
ax2.set_title('Comparação de Métricas (Precisão, Recall e Medida-F)')
ax2.set_xticks(index + bar_width)
ax2.set_xticklabels(algoritmos)
ax2.legend()

# Ajustar o layout
plt.tight_layout()

# Exibir o gráfico
plt.show()
```
